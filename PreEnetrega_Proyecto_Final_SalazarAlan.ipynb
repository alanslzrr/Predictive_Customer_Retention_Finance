- **¿Que es boxplot** El "boxplot" consiste en una caja que representa el rango intercuartílico (IQR), es decir, el rango de valores que se encuentra entre el percentil 25 (Q1) y el percentil 75 (Q3) de la distribución de la variable numérica. La línea que divide la caja en dos partes representa la mediana (o percentil 50) de la distribución.

- **Bigotes**
Los bigotes (líneas que se extienden hacia afuera de la caja) representan el rango de valores que se encuentran fuera de la caja y que no son considerados valores extremos. Estos valores se definen como aquellos que están por debajo del percentil 25 menos 1.5 veces el IQR o por encima del percentil 75 más 1.5 veces el IQR.

- **Valores Extremos**
Se representan como puntos individuales más allá de los bigotes, se definen como aquellos que están por debajo del percentil 25 menos 3 veces el IQR o por encima del percentil 75 más 3 veces el IQR.

##### C - **Income_Category** - **Label Encoding**

- Buscamos entender cuales son los valores existentes en en la variable **'Income_Category'**


````python
df['Income_Category'].value_counts(dropna=False) #Comprobacion de valores existentes, cantidades y nulls
````

*¿Porque se utilza la técnica de Label Encoding?*

- *Se llego a la conclusión de que Label Encoding es la mejor opcion, `'Income_Category'` contiene valores categóricos que no se pueden utilizar directamente en algunos algoritmos de aprendizaje automático. Por lo tanto, se convirtieron los valores categóricos en números enteros únicos. se utilizó **.remap** para  asignar manualmente un número a cada categoría para garantizar que los valores enteros asignados correspondan a la jerarquía deseada.*


----

````python
# Llenar los valores nulos con la mas frecuente
df['Income_Category'].fillna(df['Income_Category'].mode()[0], inplace=True)
````

- **fillna()**: Se utiliza **fillna()** de pandas, *Esta función reemplaza los valores nulos en la columna `'Income_Category'` con el valor de moda de esa columna. Simplemente es el valor que aparece con más frecuencia en la columna.*


````python
# Remap para el Label Encoding
remap = {
    'Less than $40K': 0,
    '$40K - $60K': 1,
    '$60K - $80K': 2,
    '$80K - $120K': 3,
    '$120K +': 4
}



# Realizar el Label Encoding con el remap
df['Income_Category'] = df['Income_Category'].map(remap)

# Convertir la columna a tipo entero
df['Income_Category'] = df['Income_Category'].astype(int)
````




- **remap** : *Se crea un diccionario llamado "remap" que mapea los valores categoricos de la columna "Income_Category" a un número entero. Esto se hace para poder hacer un "Label Encoding"  y transformar los valores categóricos en valores numéricos.*


- **astype()**: Finalmente, se convierte la columna "Income_Category" a tipo entero utilizando el método "astype()" de pandas.

#### 2.3 *Relacion Lineal*

- **Correlacion** : El método **corr()** de Pandas se utiliza para calcular la correlación entre dos columnas y devuelve un valor entre -1 y 1. Un valor de 1 indica una correlación positiva perfecta, un valor de -1 indica una correlación negativa perfecta y un valor de 0 indica que no hay correlación.

- **Covarianza**: el método **cov()** se utiliza para calcular la covarianza entre dos columnas y devuelve un valor que representa la medida en que dos variables aleatorias varían juntas. Un valor positivo indica que las variables se mueven en la misma dirección, mientras que un valor negativo indica que se mueven en direcciones opuestas.

NOTA: Aunque la correlación se define como la relación lineal entre dos variables, la presencia de una relación no lineal entre dos variables no significa que no exista correlación entre ellas. De hecho, en algunos casos, puede haber una correlación fuerte entre dos variables no lineales.

----

**¿Hay alguna relacion entre los clientes que realizan mas transacciones tambien realcien transacciones mas grandes?**

````python
# Relacion entre la cantidad de transacciones por clientes, se relaciona con el monto total de transacciones realizadas

# Calculando la correlación entre las dos columnas
correlation = df['Total_Trans_Amt'].corr(df['Total_Trans_Ct'])

# Calculando la covarianza entre las dos columnas
covariance = df['Total_Trans_Amt'].cov(df['Total_Trans_Ct'])

print("Correlación entre las dos columnas: ", correlation)
print("Covarianza entre las dos columnas: ", covariance)
````

Una **correlación alta** y una **covarianza positiva** indicarían una fuerte relación positiva entre las dos variables, lo que podría ser útil para comprender los hábitos de gasto y las necesidades de los clientes

**¿Porqué se realiza una Estandarizacion de datos?**

*La estandarización de datos es una técnica comúnmente utilizada en el preprocesamiento de datos para ajustar la escala de las variables y hacer que todas tengan la misma varianza y media, lo que ayuda a los modelos de aprendizaje automático a trabajar mejor y a mejorar la precisión de los resultados. Además, la estandarización también puede ayudar a evitar que las variables con valores atípicos dominen el modelo y afecten la precisión de los resultados.*

````python
from sklearn.preprocessing import StandardScaler

#Separacion de variable.
X = df.drop('Attrition_Flag', axis=1)
y = df['Attrition_Flag']

#Estandarizacion
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

#Matriz a Dataframe
X_std = pd.DataFrame(X_std, columns=X.columns)

#Concatenar
df_prep = pd.concat([X_std, y], axis=1)
````

1. Separar la variable objetivo (**target**) de las variables predictores.

2. Se utiliza **StandarScaler** de scikit-learn para la estandarizacion. El método **fit_transform()** de la clase StandardScaler estandariza las características de las variables predictoras y devuelve una matriz NumPy estandarizada.

3. Conversion de la matriz NumpPy a un DataFramme

4. Union de las variables predictoras **estandarizadas** con la variable objetivo (**targer**) "Attration_Flag)

En resumen, se estandarizan las variables numéricas con el dataset y se concatenan con la variable objetivo (Attrition_Flag), lo cual es necesario para que los modelos de aprendizaje automático funcionen correctamente.

**Tenemos un conjunto de datos preprocesados y estandarizados en el DataFrame**`df_prep`

**¿Porqué se dividen los datos?**

*La razón principal para hacer esto es evaluar la capacidad predictiva de un modelo sobre datos que nunca ha visto antes. Cuando un modelo se entrena en un conjunto de datos, puede memorizar los patrones y características específicas de ese conjunto de datos en lugar de aprender patrones más generales que se aplican a otros datos. Por lo tanto, para **Evaluar realmente la capacidad predictiva de un modelo**, es necesario hacer esta division.*

En esta division de datos ya estandarizados se utilza la libreria de **scikit-learn** junto a la funcion **train_test_split**. Los datos se dividiaran los datos en un **75% para entrenamiento** y un **25% para prueba**

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.25, random_state=42)
````


- **train_test_split**: *Divide los datos en conjuntos de entrenamiento y prueba.*

- **random_state=42**: *Asegura que la división sea la misma en cada ejecución del código. Los conjuntos de entrenamiento y prueba resultantes se almacenan en* `X_train`, `X_test`, `y_train`, y `y_test`.

**¿Qué es Random Forest?**

*Random forest es un algoritmo de **Clasificacion** de machine learning, combina la salida de múltiples árboles de decisión para alcanzar un solo resultado.*
*Es una técnica de ensamblaje de modelos que combina varios árboles de decisión para aumentar la precisión y reducir el riesgo de sobreajuste. En lugar de utilizar un solo árbol de decisión para la clasificación, Random Forest crea una "selva" de árboles de decisión y toma la decisión final basada en la predicción combinada de todos los árboles.*

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib
from imblearn.under_sampling import RandomUnderSampler

# RUS al conjunto de entrenamiento
rus = RandomUnderSampler(random_state=42)
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

# Entrenamiento del modelo con el conjunto de entrenamiento balanceado
clf_rf = RandomForestClassifier(random_state=42)
clf_rf.fit(X_train_rus, y_train_rus)

# Predicciones con el conjunto de prueba
y_pred_rf = clf_rf.predict(X_test)

# Guardado del modelo
joblib.dump(clf_rf, 'modelo_random_forest.p1')
````

En esta parte, se crea una instancia de **RandomUnderSampler** llamada RUS, que realiza un submuestreo aleatorio para abordar el desequilibrio de clases en el conjunto de entrenamiento. Luego, se utiliza el método **fit_resample** para aplicar el submuestreo a **X_train** y **y_train**. Los conjuntos de datos submuestreados resultantes se asignan a **X_train_rus** y **y_train_rus**.


2. *Entrenar el modelo con el conjunto de entrenamiento balanceado*

En esta parte, se crea una instancia de **RandomForestClassifier** llamada clf_rf con un generador de números aleatorios fijo utilizando **random_state=42**. Luego, se entrena el modelo utilizando el conjunto de entrenamiento balanceado **X_train_rus** y **y_train_rus** utilizando el método fit.

3. *Prediccion del modelo*

En esta parte, se utilizan los datos de prueba X_test para realizar predicciones utilizando el modelo entrenado clf_rf con el método predict. Las predicciones se asignan a y_pred_rf.

El objetivo es encontrar los coeficientes de la línea recta que minimizan la distancia entre los puntos reales y la línea predicha.
